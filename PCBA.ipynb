```python
import numpy as np
import pandas as pd
import deepchem as dc
from sklearn.ensemble import RandomForestClassifier
from deepchem.models.multitask import SingletaskToMultitask
from deepchem import metrics
from deepchem.metrics import Metric
from deepchem.models.sklearn_models import SklearnModel
from deepchem.splits import RandomStratifiedSplitter, RandomSplitter
```

```python
from deepchem.molnet.load_function.pcba_datasets import load_pcba
```

```python
tasks, datasets, transformers = dc.molnet.load_pcba(featurizer='ECFP', data_dir='./', save_dir='./')
train_dataset, valid_dataset, test_dataset = datasets
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.metrics import adjusted_rand_score
```

```python
tsne = TSNE(n_components=2, perplexity=40, random_state=104)
data_tsne = tsne.fit_transform(X)
df_tsne = pd.DataFrame(data_tsne, columns=['x_projected', 'y_projected'])
df_combo = pd.concat([dfset, df_tsne], axis=1)
sns.scatterplot(x='x_projected',y='y_projected', hue=dfset.ids, palette=sns.color_palette("husl", 10), data=df_combo)
plt.title('t-SNE Plot with Perplexity Value %s and Random State %s' %(40, 104))
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

MultitaskClassifier

```python
model = dc.models.MultitaskClassifier(n_tasks=128, n_features=1024, layer_sizes=[1000])
```

```python
metric_model = dc.metrics.Metric(dc.metrics.roc_auc_score)
print('Train set score:', model.evaluate(train_dataset, [metric_model], transformers))
```

GraphConvModel

```python
tasks, datasets, transformers = dc.molnet.load_pcba(featurizer=dc.feat.ConvMolFeaturizer())
train_dataset, valid_dataset, test_dataset = datasets
```

```python
model = dc.models.GraphConvModel(n_tasks=128, mode='classification')
```

```python
metric = dc.metrics.Metric(dc.metrics.roc_auc_score)
print('Train set score:', model.evaluate(train_dataset, [metric], transformers))
print('Test set score:', model.evaluate(test_dataset, [metric], transformers))
```

GATmodel

```python
tasks, datasets, transformers = dc.molnet.load_pcba(featurizer=dc.feat.MolGraphConvFeaturizer())
train_dataset, valid_dataset, test_dataset = datasets
```

```python
from deepchem.models import GATModel
model = dc.models.GATModel(mode='classification', n_tasks=128,batch_size=1024, learning_rate=0.001)
```

```python
metric = dc.metrics.Metric(dc.metrics.roc_auc_score)
print('Train set score:', model.evaluate(train_dataset, [metric], transformers))
print('Test set score:', model.evaluate(test_dataset, [metric], transformers))
```
